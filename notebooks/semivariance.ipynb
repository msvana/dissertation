# Semivariance as a metric for determining TFN support

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
base_data = np.zeros(20)
```

```python
def semivariance(data, side):
    mean = np.mean(data)
    if side == 'left':
        semi_data = data[data < mean]
    else:
        semi_data = data[data > mean]
    result = np.sum((semi_data - mean)**2) / len(semi_data)
    return result
```

```python
def calculate_and_draw(data):
    mean = data.mean()
    semi_left = semivariance(data, 'left')**(1/2)
    semi_right = semivariance(data, 'right')**(1/2)
    coef = 1.5

    plt.hist(data, bins=len(data), rwidth=0.5)
    plt.plot([mean-semi_left*coef, mean, mean+semi_right*coef], [0, 1, 0])
    plt.plot([-1, -0.5, 0], [1, 1, 0])
    plt.plot([0, 0.5, 1], [0, 1, 1])
    plt.show()
    
uniform = np.arange(-1, 1.1, 0.1)
calculate_and_draw(uniform)
```

```python
uniform_left = np.arange(-1, 0.1, 0.1)
calculate_and_draw(uniform_left)
```

```python
uniform_mid = np.arange(-0.5, 0.6, 0.1)
calculate_and_draw(uniform_mid)
```

```python
uniform_shifted = np.arange(-0.7, 0.4, 0.1)
calculate_and_draw(uniform_shifted)
```

```python
calculate_and_draw(np.array([-1, -1, -1, 0.5, 0.5, 0.5]))
```

```python
calculate_and_draw(np.array([-1, -1, -1, 1, 1, 1, 0.1]))
```

```python
calculate_and_draw(np.array([-1, 1]))
```
