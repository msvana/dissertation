```python
import matplotlib.pyplot as plt
import pandas as pd


plt.rcParams.update({'font.size': 15})
```

```python
df = pd.read_csv('../datasets/topics_brno_processed.csv')
top_topics = df.sort_values('mass', ascending=False)[:8]
```

```python
plt.figure(figsize=(15, 8))
plt.xlabel('Polarity')
plt.plot([-1, 0.75, 1], [1, 0, 0], color='red', label='Negative opinion', linestyle="--")
plt.plot([-1, -0.32, 1], [0, 0, 1], color='green', label='Positive opinion', linestyle="--")

for _, topic in top_topics.iterrows():
    plt.plot([topic['support_left'], topic['sent_mean'], topic['support_right']], [0, 1, 0], label=topic['name'])

plt.legend(loc="upper left")
plt.show()
```
